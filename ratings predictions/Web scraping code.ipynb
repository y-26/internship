```python
#Importing required modules
import pandas as pd
import numpy as np
import selenium
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException

#Details to collect
links=[]
```

```python
product=['phones','laptop','headphones','printers','watches']

for i in product:
    #Initialising the driver
    driver=webdriver.Chrome(r"I:\\chromedriver.exe")

    #open the required website
    driver.get("https://www.flipkart.com/")
    time.sleep(3)

    #Closing the popup
    try:
        popup=driver.find_element_by_xpath("//button[@class='_2KpZ6l _2doB4z']")
        popup.click()
    except:
        pass

    #Finding the search bar in website
    search_bar=driver.find_element_by_name('q')
    search_bar.send_keys(i)

    #Finding and Clicking on the search button
    search_button=driver.find_element_by_class_name("L0Z3Pu")
    search_button.click()
    time.sleep(5)

        
    #Getting the links of all products
    if i=='phones' or i=='laptop':
        productlinks=driver.find_elements_by_xpath('//a[@class="_1fQZEK"]')
        for i in productlinks:
            links.append(i.get_attribute('href'))

    elif i == 'printers' or i=='headphones':
        productlinks=driver.find_elements_by_xpath('//a[@class="s1Q9rs"]')
        for i in productlinks:
            links.append(i.get_attribute('href'))

    else:
        productlinks=driver.find_elements_by_xpath('//a[@class="_2UzuFa"]')
        for i in productlinks:
            links.append(i.get_attribute('href'))
                
        
    driver.close()

print(len(links))
```

```python
review=[]
rating=[]
d_review=[]
#Initialising the driver
driver=webdriver.Chrome(r"I:\\chromedriver.exe")
    
for link in links:
    driver.get(link)
    time.sleep(5)
    
    #All reviews
    try:
        driver.execute_script("window.scrollTo(0,document.body.scrollHeight/2)")
        all_re=driver.find_element_by_xpath('//div[@class="_3UAT2v _16PBlm"]').click()
        time.sleep(3)
    except NoSuchElementException:
        pass
    
    #pages
    pge=[]
    try:
        review_pages=driver.find_elements_by_xpath('//nav[@class="yFHi8N"]/a')
        for i in review_pages:
            pge.append(i.get_attribute('href'))
    except NoSuchElementException:
        pass
    
    length=len(pge)
    
    if length > 30:
        pge=pge[:30]
    else:
        pge=pge
        
    for pges in pge:
        driver.get(pges)
        time.sleep(5)
        
        
        #Ratings
        ratings=driver.find_elements_by_xpath('//div[@class="col _2wzgFH K0kLPL"]/div[1]/div[1]')
        for r in ratings:
            rating.append(r.text)

        #Review
        reviews=driver.find_elements_by_xpath('//p[@class="_2-N8zT"]')
        for i in reviews:
            review.append(i.text.replace('\n',''))
            
        #Detailed Review
        detailed_reviews=driver.find_elements_by_xpath('//div[@class="t-ZTKy"]/div[1]/div[1]')
        for i in detailed_reviews:
            d_review.append(i.text.replace('\n',''))
               
print(len(rating), len(review), len(d_review))        
print(rating)
review
```

```python
data=pd.DataFrame({})
data['Ratings']=rating
data['Reviews']=review
data['Detailed_Review']=d_review
data
```
